```python
#讀檔後斷詞，並統計tf及df
#Big Data & Business Analytics, National Taiwan University
```

```python
import csv
import monpa
from monpa import utils
from collections import Counter

tf_counter=Counter() #預備統計tf用
df_counter=Counter() #預備統計df用

f = open("../../bda2023_mid_dataset/stock_data_2019-2023.xlsx", newline='', encoding='utf-8') #開檔 
rows = csv.reader(f, delimiter=',') #讀檔
i=0
for row in rows: #對內容逐行列舉，row[0]是title, row[1]是content
  i=i+1 
  print(i)
  if(i>1): #第一行是欄位名稱，掠過
        
    #以下套用斷詞小範例
    query=row[0]+' '+row[1]
    df_tmp=Counter() #暫存本篇df用
    sentence_list = utils.short_sentence(query) #斷句
    for item in sentence_list:
      result_cut = monpa.cut(item) #斷詞
      for term in result_cut:
        term=term.strip() #去除前後多餘空白
        if(len(term)>1): #若詞長>1
          tf_counter[term]+=1 #tf加1
          if(df_tmp[term]==0): #若本篇之前不曾出現
            df_tmp[term]=1 #df標為1
    df_counter+=df_tmp # 累加多篇df

f.close() #關閉檔案
```

```python
for term in tf_counter.most_common(100): #印出tf前100名
  print(term[0],term[1])
```

```python
for term in df_counter.most_common(100): #印出df前100名
  print(term[0],term[1])
```

```python
f = open('d:/tmp/limit_down_term.csv', 'w', newline='', encoding='utf-8') #以寫入模式開啟另一個檔
writer = csv.writer(f) #準備寫入
for term in tf_counter.most_common(100): #印出tf前100名
  writer.writerow(term) #寫入一行
f.close()
```
