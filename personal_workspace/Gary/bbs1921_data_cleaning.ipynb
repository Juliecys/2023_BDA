```python
import pandas as pd
import numpy as np
import datetime

import warnings
warnings.filterwarnings("ignore")
```

```python
#  讀檔
bbs1921 = pd.read_csv("../../bda2023_mid_dataset/bda2023_mid_bbs_2019-2021.csv")
bbs1921
```

```python
#  篩選出和半導體有關的文章
bbs1921_wafer = bbs1921[bbs1921['title'].str.contains("台積電|張忠謀|晶片|晶圓|封裝|聯電|聯發科|聯詠") | bbs1921['content'].str.contains('台積電|張忠謀|晶片|晶圓|封裝|聯電|聯發科|聯詠')].reset_index(drop = True)
bbs1921_wafer
```

```python
#  轉換日期型別
bbs1921_wafer['post_time'] = pd.to_datetime(bbs1921_wafer['post_time']).dt.date
bbs1921_wafer
```

```python
#  保留文章日期、標題、內文
bbs1921_wafer = bbs1921_wafer[['post_time', 'title', 'content']]
bbs1921_wafer
```

```python
#  斷詞套件
import monpa
from monpa import utils
import re
```

```python
#  用來將字串以正則化處理去除中文字元以外的字元
def clearSentence(sentence):
    return re.sub(r'[^\u4e00-\u9fa5]+', '', sentence)

#  去除繁體中文常用詞
with open('C:/Users/ASUS/OneDrive/桌面/stopwords_zh.txt', 'r', encoding='utf-8') as file:
    stopwords = file.read().splitlines()
file.close()
stopwords
```

```python
train_startDate = datetime.date(2019,1,1)
train_endDate = datetime.date(2021,12,31)

#  透過monpa對文章進行斷詞處理，並將個別斷詞結果寫在tokenStr這個字串變數中，
#  最後再將訓練集中所有字串存在tokenStr_list中
train_tokenStr_list = list()
for i in list(bbs1921_wafer[bbs1921_wafer["post_time"].between(train_startDate, train_endDate)].index):
    try:
        sentence_list = utils.short_sentence(bbs1921_wafer['content'][i])
        tokenStr = str()
        for sentence in sentence_list:
            sentence = clearSentence(sentence)
            tokens = monpa.cut(sentence)
            tokenStr += ' '.join(tokens)
        train_tokenStr_list.append(tokenStr)
    except:
        train_tokenStr_list.append('')
len(train_tokenStr_list)
```
